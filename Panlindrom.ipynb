```python
import random
from itertools import combinations

arr  = []
arr_1 = []
arr_sort = []

for n1 in range(101,999):
  arr_1.append(n1)

for i in range(0, len(arr_1),1):
  for j in range(0,len(arr_1),1):
    arr_2 = arr_1[i]*arr_1[j]
    if len(str(arr_2))>5 and not str(arr_1[i]).endswith('0'):
      arr.append(arr_2)

for i in range(0, len(arr)):
  if str(arr[i]) == str(arr[i])[::-1]:
    arr_sort.append(str(arr[i]))
    # print(str(arr[i])[::-1])
arr_sort.sort(reverse=True)


```
